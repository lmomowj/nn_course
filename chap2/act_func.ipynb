### 定义激活函数

####  阈值型

&nbsp; $\cdot $&nbsp; 单位阶跃函数&nbsp; $u(z) = 
\begin{cases}
0 & z<0 \\ 
1 & z \geq 0
\end{cases}
$

&nbsp; $\cdot $&nbsp; 符号函数 &nbsp; $\mathrm{sgn}(z) = 
\begin{cases}
-1 & z<0 \\
0 & z=0 \\ 
1 & z > 0
\end{cases}$

#### Sigmoid型

&nbsp; $\cdot $&nbsp; Logistic函数&nbsp; $\sigma(z) = \frac{1}{1+\mathrm{e}^{-z}}$

&nbsp; $\cdot $&nbsp; Tanh 双曲正切函数 &nbsp; $\mathrm{tanh}(z) = \frac{\mathrm{e}^z-\mathrm{e}^{-z}}{\mathrm{e}^z+\mathrm{e}^{-z}}$

&nbsp; $\cdot $&nbsp; 双极性Sigmoid函数&nbsp; $f(z) = \frac{1-\mathrm{e}^{-z}}{1+\mathrm{e}^{-z}}$

#### 分段线性

&nbsp; $\cdot $&nbsp; 单极性分段线性变换函数 &nbsp; $f(z) = 
\begin{cases}
0 & z \leq 0 \\
cz & 0 < z \leq z_c \\ 
1 & z > z_c
\end{cases}$

&nbsp; $\cdot $&nbsp; 双极性分段线性变换函数 &nbsp; $f(z) = 
\begin{cases}
-1 & z \leq 0 \\
cz-1 & 0 < z \leq z_c \\ 
1 & z > z_c
\end{cases}$

&nbsp; $\cdot $&nbsp; Hard-Logistic 函数 &nbsp; $f(z) = 
\begin{cases}
0 & g_l(z) \leq 0 \\
g_l(z) & 0 < g_l(z) < 1 \\ 
1 & g_l(z) \geq 1
\end{cases} =  \max(\min(g_l(z),1),0) =  \max(\min(0.25z+0.5,1),0)$

&nbsp; $\cdot $&nbsp; Hard-Tanh 函数 &nbsp; $f(z) = 
\begin{cases}
0 & g_l(z) \leq -1 \\
g_l(z) & -1 < g_l(z) < 1 \\ 
1 & g_l(z) \geq 1
\end{cases} = \max(\min(g_l(z),1),-1) 
= \max(\min(z,1),-1) $

&nbsp; $\cdot $&nbsp; ReLU（修正线性单元 rectified linear unit） &nbsp; $\mathrm{ReLU}(z) = \max(0,z)
= \begin{cases}
z & z \geq 0 \\
0 & z<0
\end{cases}$

&nbsp; $\cdot $&nbsp; Leaky ReLU（带泄露的ReLU） &nbsp; 
$\mathrm{LeakyReLU}(z) = \max(z,\alpha z)
=\begin{cases}
z & z > 0 \\
\alpha z & z \leq 0
\end{cases}
\ \ (0<\alpha<1)$

&nbsp; $\cdot $&nbsp; ELU (Exponential Linear Unit  指数线性单元) &nbsp; 
$\mathrm{ELU}(z) 
= \begin{cases}
z & z \geq 0 \\
\alpha(\mathrm{e}^z-1) & z<0
\end{cases}$

&nbsp; $\cdot $&nbsp; softolus 函数 &nbsp; 
$\mathrm{softplus}(z) = \log(1 + \mathrm{e}^z)$

&nbsp; $\cdot $&nbsp; Swish 函数 &nbsp; 
$\mathrm{swish}(z) = z\sigma(\beta z) = \frac{z}{1+\mathrm{e}^{-\beta z}}$

&nbsp; $\cdot $&nbsp; GELU (Gaussian Error Linear Unit, 高斯误差线性单元) &nbsp; 
$\mathrm{GELU}(z) = z P(Z\leq z)$  &nbsp;  &nbsp;  &nbsp; $P(Z\leq z)$ 是Gaussian分布 $\mathcal{N}(\mu,\sigma^2)$ 的累积分布函数 

### 函数求导

```python
print('激活函数：{}'.format(', '.join(act.func_name.keys())))
# 函数名称
fname = 'relu'  #  'gelu'    #   'GELU' #  'Tanh'   #  'Logistic'
# 调用函数
func = eval(f'act.{act.func_name[fname]}()')
print("{} 函数: {}".format(fname,func.formula()))
x = 0
value = func.value(x)
print('f({}) = {}'.format(x,value))
# 计算导数
gradient = func.gradient(order=1)
print("{} 函数的导数: {}".format(fname,gradient))
df = func.gradient_value(x,order=1)
print('df({}) = {}'.format(x,df))
```
